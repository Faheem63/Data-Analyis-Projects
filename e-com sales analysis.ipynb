```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
data = pd.read_csv("Superstore.csv", encoding='latin-1') 
#encoding='latin-1': Special characters ko properly read karne ke liye encoding use hui hai
data.head()
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date']) 
#Date Conversion: Order Date aur Ship Date columns ko datetime format me convert kiya gaya hai for date-based analysis.
```

```python
data['Order Month'] = data['Order Date'].dt.month 
data['Order Year'] = data['Order Date'].dt.year
data['Order Day of Week'] = data['Order Date'].dt.dayofweek

#Order Month: Order date se month extract karte hain.
#Order Year: Order date se year extract hota hai.
#Order Day of Week: Week ka day (0 for Monday, 6 for Sunday) extract kiya gaya hai.
```

```python
data.head()
```

```python
# 1. MONTHLY SALES ANALYSIS

monthly_sales = data.groupby("Order Month")["Sales"].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(data=monthly_sales, x="Order Month", y="Sales", marker="o")
plt.title("Monthly Sales Analysis")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
# 2. YEARLY SALES ANALYSIS

yearly_sales = data.groupby("Order Year")["Sales"].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(data=yearly_sales, x="Order Year", y="Sales")
plt.title("Yearly Sales")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
# 3. SALES BY CATEGORY (BAR CHART)

category_sales = data.groupby("Category")["Sales"].sum().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=category_sales, x="Category", y="Sales")
plt.title("Category-wise Sales")
plt.show()
```

```python
# 4. SALES BY SUB-CATEGORY

subcat_sales = data.groupby("Sub-Category")["Sales"].sum().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(data=subcat_sales, x="Sub-Category", y="Sales")
plt.title("Sub-Category Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. PROFIT BY CATEGORY

category_profit = data.groupby("Category")["Profit"].sum().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=category_profit, x="Category", y="Profit")
plt.title("Category-wise Profit")
plt.show()
```

```python
# 6. PROFIT BY SUB-CATEGORY (SORTED)

subcat_profit = (
    data.groupby("Sub-Category")["Profit"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

plt.figure(figsize=(12, 6))
sns.barplot(data=subcat_profit, x="Sub-Category", y="Profit")
plt.title("Profit by Sub-Category")
plt.xticks(rotation=45)
plt.show()
```

```python
# 7. REGION-WISE SALES

region_sales = data.groupby("Region")["Sales"].sum().reset_index()

plt.figure(figsize=(8, 5))
sns.barplot(data=region_sales, x="Region", y="Sales")
plt.title("Sales by Region")
plt.show()
```

```python
# 8. SHIP MODE ANALYSIS

ship_sales = data.groupby("Ship Mode")["Sales"].sum().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(data=ship_sales, x="Ship Mode", y="Sales")
plt.title("Sales by Shipping Mode")
plt.show()
```

```python
# 9. CORRELATION HEATMAP

plt.figure(figsize=(10, 6))
sns.heatmap(data[['Sales', 'Quantity', 'Discount', 'Profit']].corr(),
            annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
